Train test split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
X=df.drop('prognosis',axis=1)
```

```python
y=df['prognosis']
```

```python
le = LabelEncoder()
```

```python
le.fit(y)
Y=le.transform(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=20)
```

Training top models

```python
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

storing models in dictionary

```python
models={"RF":RandomForestClassifier(n_estimators=100,random_state=42),
        "GBC":GradientBoostingClassifier(n_estimators=100,random_state=42),"SVC":SVC(kernel='linear'),"KNN":KNeighborsClassifier(n_neighbors=5),"MNB":MultinomialNB()}
```

```python
for models_name,model in models.items():
    #training model
    
    model.fit(X_train,y_train)
    
    #testing model
    
    pre=model.predict(X_test)
    
    #checking the accuracy
    
    accuracy=accuracy_score(y_test,pre)
    
    #Confusion matrix
    
    cm= confusion_matrix(y_test,pre)
    
    print(f"""{models_name} accuracy : {accuracy}
        {models_name} Confusion Matrix \n{np.array2string(cm,separator=',')}\n""")
```

Solo prediction

```python
svc=SVC(kernel='linear')
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
accuracy_score(y_test,y_pred)
```

saving the model in models folder

```python
import pickle
pickle.dump(svc,open("models/svc.pkl",'wb'))
```
